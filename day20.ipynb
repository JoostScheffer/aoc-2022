```python
A = np.loadtxt("in20.txt", dtype=int)
L = len(A)
B = np.vstack((A, np.arange(L))).T


print(np.array(sorted(B.tolist(), key=lambda x: x[1]))[:, 0])

for n, (num, idx) in enumerate(B):
    idx = B[n, 1]

    new_idx = idx + num
    if new_idx <= 0:
        # it goes down
        new_idx = (L - 1) + new_idx
    else:
        # it goes up
        new_idx %= L

#     print("will move down", B[(B[:, 1] > idx) * (B[:, 1] <= new_idx), 0])
#     print("will move up  ", B[(B[:, 1] <= idx + 1) * (B[:, 1] > new_idx + 1), 0])
#     print(num, ": ", idx, "->", new_idx)

    B[(B[:, 1] > idx) * (B[:, 1] <= new_idx), 1] -= 1
    B[(B[:, 1] <= idx + 1) * (B[:, 1] > new_idx + 1), 1] += 1
    B[n, 1] = new_idx

#     print(np.array(sorted(B.tolist(), key=lambda x: x[1]))[:, 0])
#     print(sorted(B[:, 1]), "all idxs")
#     print()


# print(sorted(B[:, 1]))
res = np.array(sorted(B.tolist(), key=lambda x: x[1]))[:, 0]
print(res)
```
